# This notebook shows the usage of plotting.lineplot and report.Html

```python
import sys
sys.path.insert(0, "../../..")
import backtest
from backtest.lib import plotting, report
import pandas as pd
import numpy as np
%load_ext autoreload
%autoreload 2
```

```python
df = pd.DataFrame(np.random.randn(10, 3), columns=["y1", "y2", "y3"])
df["id"] = range(10)
df.head()
```

```python
g = plotting.lineplot(df, 
    x="id", 
    y=["y1", "y2", "y3"],
    title="ttttt",
    xlab="xxxx",
    ylab="yyyy",
    transform=np.cumsum,
    stats="mean")
```

```python
h = report.Html(out_dir="../../data/html_test",
    name="test")
h.add_table(df, name="random table")
h.add_figure(g, name="cumsum ")
h.compile()
```
